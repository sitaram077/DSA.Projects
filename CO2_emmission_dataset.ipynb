```python
plt.scatter(train.EngineSize,train.CO2emission)
plt.plot(x_train,regr.coef_[0][0]*x_train + regr.intercept_[0],'-y')


```

# MULTIPLE LINEAR REGRESSION

```python
split_df=df[['Engine Size(L)','Cylinders','Fuel Consumption Comb (mpg)','CO2 Emissions(g/km)']]
moddy_df=split_df.rename (columns={'Fuel Consumption Comb (mpg)':'FuelConsumption'})
moddy2_df=moddy_df.rename(columns={'Engine Size(L)':'EngineSize'})
moddy3_df=moddy2_df.rename(columns={'CO2 Emissions(g/km)':'CO2emission'})
moddy3_df
```

```python
plt.scatter(moddy3_df.EngineSize,moddy3_df.CO2emission,color='red')
plt.xlabel('enginesize')
plt.ylabel('co2emission')
plt.show()
```

```python
mask=np.random.rand(len(moddy3_df))<0.8
mlr_train=moddy3_df[mask]
mlr_test=moddy3_df[~mask]

```

```python
from sklearn import linear_model
mul_reg=linear_model.LinearRegression()
x_ax=np.asanyarray(mlr_train[['EngineSize',	'Cylinders','FuelConsumption']])
y_ax=np.asanyarray(mlr_train[['CO2emission']])
mul_reg.fit(x_ax,y_ax)
print('cooeficeint:',mul_reg.coef_)
```

```python
y_hat=mul_reg.predict(mlr_train[['EngineSize',	'Cylinders','FuelConsumption']])
print("residual sum of scores:%.2f"%np.mean((y_hat-y_ax)**2))
print("score",mul_reg.score(x_ax,y_ax))
```
